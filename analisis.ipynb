# Análisis del Motor de Búsqueda Semántica

Visualización de resultados del motor de búsqueda usando embeddings de OpenAI y ChromaDB.

```python
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from chromadb.utils.embedding_functions import DefaultEmbeddingFunction

sys.path.append('.')
from buscar import buscar

print('Setup completo')
```

## Parte 1: Queries de prueba y sus resultados

```python
queries = [
    "¿cómo hacer una API en Python?",
    "diferencias entre frameworks de frontend",
    "cómo funciona la autenticación en aplicaciones web",
    "herramientas para trabajar con modelos de lenguaje",
]

resultados_por_query = {}

for query in queries:
    resultados = buscar(query, n_resultados=3)
    resultados_por_query[query] = resultados
    print(f"\nQuery: {query}")
    print("-" * 60)
    for r in resultados:
        print(f"  [{r['score']:.4f}] {r['titulo']}")
        print(f"           {r['contenido'][:80]}...")
```

## Parte 2: Tabla comparativa

```python
filas = []
for query, resultados in resultados_por_query.items():
    mejor = resultados[0]
    filas.append({
        "Query": query,
        "Mejor resultado": mejor["titulo"],
        "Score de similitud": round(mejor["score"], 4),
        "2º resultado": resultados[1]["titulo"] if len(resultados) > 1 else "-",
        "3º resultado": resultados[2]["titulo"] if len(resultados) > 2 else "-",
    })

df = pd.DataFrame(filas)
df.style.set_caption("Resultados del motor de búsqueda semántica") \
        .background_gradient(subset=["Score de similitud"], cmap="Greens") \
        .set_properties(**{"text-align": "left"})
```

## Bonus: Mapa de calor de similitud entre artículos

```python
from indexar import articulos

ef = DefaultEmbeddingFunction()
contenidos = [art['contenido'] for art in articulos]
titulos = [art['titulo'] for art in articulos]

embeddings = np.array(ef(contenidos))

norms = np.linalg.norm(embeddings, axis=1, keepdims=True)
embeddings_norm = embeddings / norms
similitud = np.dot(embeddings_norm, embeddings_norm.T)

print(f'Matriz de similitud: {similitud.shape}')
print(f'Rango de valores: [{similitud.min():.3f}, {similitud.max():.3f}]')
```

```python
titulos_cortos = [
    t.replace(" para ", "\npara ").replace(" vs ", "\nvs ").replace(" con ", "\ncon ")
    for t in titulos
]

fig, ax = plt.subplots(figsize=(12, 10))

mask = np.eye(len(articulos), dtype=bool)  # ocultar diagonal (siempre = 1)

sns.heatmap(
    similitud,
    annot=True,
    fmt=".2f",
    cmap="YlOrRd",
    xticklabels=titulos_cortos,
    yticklabels=titulos_cortos,
    vmin=0,
    vmax=1,
    linewidths=0.5,
    ax=ax
)

ax.set_title("Similitud coseno entre artículos", fontsize=14, pad=15)
plt.xticks(fontsize=8, rotation=45, ha="right")
plt.yticks(fontsize=8, rotation=0)
plt.tight_layout()
plt.savefig("similitud_articulos.png", dpi=150, bbox_inches="tight")
plt.show()
print("Heatmap guardado como similitud_articulos.png")
```
